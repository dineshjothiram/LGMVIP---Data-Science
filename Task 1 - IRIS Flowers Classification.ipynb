### Optimal Cluster Identification

```python
from sklearn.cluster import KMeans

n_clus = list(range(1,12))

interia_list = []

for i in n_clus:
    model = KMeans(n_clusters = i)
    model.fit(iris_input)
        
    interia_list.append(model.inertia_)
    
print("No of clusters\n",n_clus)

print("\n Inertia List\n",interia_list)
```

#### The above code is used to represent the number of cluster and its corresponding Intertia values list. The code represent the 11 clusters and its corresponding intertia values.

```python
plt.plot(n_clus,interia_list,marker='s',markersize= 4 ,c='green')
plt.xlabel('Number of Clusters')
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

The above graph is represent the Intertia and Number of clusters which is known as Elbow method.When the number of cluster is 1 the intertia value is very high. When the cluster increased by 1 i.e number of cluster is 2, the inertia value completely decreased.
When the number of cluster value is 3 the intertia values not varies so much. From the observation we can conclude that the best or optimal number of cluster is 3.

### KMean Clustering Algorithm

```python
from sklearn.cluster import KMeans
km_model = KMeans(n_clusters= 3)
km_model.fit(iris_input)
```

The above code we used KMeans clustering algorithm with number of cluster values is 3. Then we have to fit out dataset into KMeans algorithm.

```python
#Print the cluster centers for each cluster
km_model.cluster_centers_
```

```python
print(len(km_model.cluster_centers_))
```

```python
#Print the Inertia value for 3 cluster
km_model.inertia_
```

### Plotting the Clusters

```python
plt.figure(figsize=(9, 6))
plt.scatter(x = iris_input['sepal length (cm)'], y =iris_input['sepal width (cm)'], c=km_model.labels_)
plt.scatter(x = iris_input['petal length (cm)'], y =iris_input['petal width (cm)'], c=km_model.labels_)
plt.show()
```

#### The above diagram shown 3 clusters with 3 different colors of the IRIS dataset. 

### Plotting Clusters with Centroid values

```python
plt.figure(figsize=(9, 6))
plt.scatter(x=iris_input['sepal length (cm)'], y=iris_input['sepal width (cm)'], c = km_model.labels_,  s = 54)
plt.scatter(x=iris_input['petal length (cm)'], y=iris_input['petal width (cm)'], c = km_model.labels_,  s = 45)
plt.scatter(x=km_model.cluster_centers_[:,0], y=km_model.cluster_centers_[:,1], c = 'red', marker = '*', s= 200)
plt.show()
```

In the above digram the RED color STAR points represents the centroid value of the each cluster.

### Performace calculation  of KMeans Clustering Algorithm

```python
from sklearn import metrics
score = metrics.silhouette_score(X=iris_input,labels=km_model.labels_)
print("Silhouette Score is",score)
```

#### Silhouette Score is used to calculate the performace of KMeans clustering algorithm. Our Silhouette Score is 0.55 for 3 clusters for the IRIS dataset.
